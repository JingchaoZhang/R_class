```python
data = pd.read_csv('us-counties.csv')
```

```python
decatur = data[(data['county'] == 'Decatur') & (data['state'] == 'Georgia')].set_index('date')['cases']
decatur
```

```python
Lincoln = data[(data['county'] == 'Lincoln') & (data['state'] == 'Nebraska')].set_index('date')['cases']
Lincoln
```

```python
decatur.plot()
Lincoln.plot()

plt.legend(['Decatur, GA', 'Lincoln, NE'])

plt.gca().set_title('Cumulative covid19 cases in Decatur, GA and Lincoln, NE', pad = 20)
plt.gca().set_ylabel('Number of cases')
plt.gca().set_xlabel('Date')

plt.savefig('covid19.png')
```
